```python
# Imports
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
from errno import EFBIG


G = nx.random_graphs.extended_barabasi_albert_graph(50, 1, 0.2, 0.1,seed=10)

klists = [[10, 14, 3], [10, 14, 3, 4, 2], [10, 14, 3, 4, 2, 31, 49], [10, 14, 3, 4, 2, 31, 49, 21, 25, 36, 43]]
kin = klists[0]
```

```python
for kin in klists:
    # Implement steiner_tree function
    Gstein = nx.algorithms.approximation.steinertree.steiner_tree(G, kin)
    color_map = []
    for node in Gstein:
        if node in kin:
            color_map.append('red')
        else:
            color_map.append('green')

    nx.draw(Gstein, with_labels=True, font_color='white', node_color=color_map)
    plt.savefig(f'Q-4.2.1.Steiner_tree_nodes_{kin}.png')
    nx.write_adjlist(Gstein, f'Q-4.2.1.Steiner_tree_nodes_{kin}.npy')
```
